```python
#load car data in pandas dataframe
quality=pandas.read_csv('C:/Users/ACER/Desktop/packs/car.csv')
```

```python
#taking dependent and indepedent variables seperately
quality_X=quality[['buying','maint','doors','persons','lug_boot','safety']]
quality_Y=quality[['class values']]
```

```python
#divide the data into train and test set.... train data set contains 80% of data and set contains 20% of data.
X_train,X_test,Y_train,Y_test=train_test_split(quality_X,quality_Y,test_size=0.20)
```

```python
#assigning different values for different fields in buying column.
X_train['buying'] = X_train['buying'].map({'low': 0,'med': 1,'high':2,'vhigh':3})
```
